```python
from matplotlib import pyplot as plt
import os
import pickle
import numpy as np
```

```python
path = './adv_data/first_frame/0.5_008240.pickle'


#files = os.listdir(path)
#files.sort()
#for file in files:
#    if '.pickle' not in file:
#        continue
with open(os.path.join(path),'rb') as fr:
    datas = pickle.load(fr)

try:
    org,adv = datas['examples'][0]
    org = (((org + 1)/2)*255).astype('int32')
    adv = (((adv + 1)/2)*255).astype('int32')
    org, adv = np.squeeze(org[:,0,:,:]), np.squeeze(adv[:,0,:,:])
    org, adv = np.transpose(org, (1,2,0)), np.transpose(adv, (1,2,0))
except Exception as exc:
    raise RuntimeError("adv examples are all correct") from exc


# 3 x 64 x 224 x 224

plt.figure(figsize=(8,10))
plt.subplot(1,2,1)
plt.title("original")
plt.ylabel("Eps: {}".format(datas['epsilon']), fontsize=14)
plt.xticks([], [])
plt.yticks([], [])
plt.imshow(org)
plt.subplot(1,2,2)
plt.title("perturbed")
plt.xticks([], [])
plt.yticks([], [])
plt.imshow(adv)
plt.show()
```

```python
epsilons = [0, 0.1, 0.5, 1, 2, 4, 8]
path = './adv_data/all/{}_008240.pickle'

print('attack on every frame')
print()

for i in epsilons:
    with open(path.format(i),'rb') as fr:
        datas = pickle.load(fr)
        print("epsilon: %.1f    accuracy_org: %.8f     accuracy_adv: %.8f" 
              % (datas['epsilon'], datas['accuracy_org'], datas['accuracy_adv']))
```
